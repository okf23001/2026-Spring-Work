```python
from PIL import Image
import os

def convert_image(input_path, output_path):
    img = Image.open(input_path)
    img = img.convert('L')
    img = img.resize((28, 28), Image.LANCZOS)
    #Use below line for color inversion
    img = Image.eval(img, lambda x: 255 - x)
    #Use below line for black and white
    img = img.point(lambda x: 255 if x >= 128 else 0, '1')
    img.save(output_path)
    print(f"Saved: {output_path}")

#        Convert Folder
input_folder = '/Users/owenfinch/Downloads/Joo Group Work/SS Test Photos'
output_folder = '/Users/owenfinch/Downloads/Joo Group Work/Edited Test'
os.makedirs(output_folder, exist_ok=True)

for filename in os.listdir(input_folder):
    if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
        input_path = os.path.join(input_folder, filename)
        output_path = os.path.join(output_folder, f"converted_{filename}")
        convert_image(input_path, output_path)
```
